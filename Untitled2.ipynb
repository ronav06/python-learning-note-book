```python
from sklearn.linear_model import LinearRegression
import numpy as np

time = np.array([1, 2, 3, 4]).reshape(-1, 1)
marks = np.array([10, 20, 30, 40])

model = LinearRegression()
model.fit(time, marks)

time_used = float(input("enter the time"))

print(model.predict([[time_used]]))
print("the exam marks are most likely to be:", model.predict([[time_used]]))
```

```python
temperature = [10,20,35,65,87,1,34]
days = ["monday","tuesday","wednesday","thursday","friday","saturday","sunday"]

highest_temperature = max(temperature)
lowest_temperature = min(temperature)

print("the highest temperature is:",highest_temperature,"which is on",days[temperature.index(highest_temperature)])
print("the lowest temperature is:",lowest_temperature ,"which is on",days[temperature.index(lowest_temperature)])
```

# line graph

```python
import matplotlib.pyplot as plt

days =["monday","tuesdday","wednesday","thursday","friday","saturday","sunday"]
temperature = [12, 23,14,54,33,65,21]
plt.plot(days, temperature, marker=0)
plt.title("temperature of the week")
plt.xlabel("days of the week")
plt.ylabel("temperature")
plt.show()
```

# bar graph

```python
import matplotlib.pyplot as plt

days = ["monday","tuesday","wednesday","thursday","friday","saturday","sunday"]
temperature = [12,23,14,54,33,65,21]
plt.bar(days, temperature)
plt.title("temperature of the week")
plt.xlabel("days of the week")
plt.ylabel("temperature")
plt.show()
```
